```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import minmax_scale
%matplotlib inline
from skcriteria import Data, MIN, MAX
import numpy as np
```

```python
df = pd.read_excel (r'STR-Ene-22.xlsx')
```

```python
Manual=[]
Campanas=[]
for i in range(len(df)):
    u=df.iloc[i][4]
    x=df.iloc[i][7]
    y=df.iloc[i][6]
    if u in Campanas:
        Y="SB"
        Manual.append(Y)
        
    else:
        if y=="*":

            Y="Automatica"
            Manual.append(Y)


        else:
            Y="Manual"
            Manual.append(Y)

    
df["Tipo"]=Manual

temp=df
```

```python
targeting=[]

for i in range(len(df)):
    y=df.iloc[i][6]
    x=df.iloc[i][6]
    u=df.iloc[i][7] 
    p=df.iloc[i][7] 
    
    if u =="EXACT" or u=="BROAD" or u=="PHRASE":    
        if x=="*":
            u=df.iloc[i][8]
            targeting.append(u)
        elif x!="*":
            u=df.iloc[i][6]
            targeting.append(u)
        
    elif y[0:4]=="asin":
        y=y[6:]
        y=y[:-1]
        y=y.lower()
        targeting.append(y)
          
    elif x=="*" and p !="BROAD":
        x=df.iloc[i][8]
        targeting.append(x)
        
    else:
        targeting.append(df.iloc[i][6])
        
        
df["Targeting1"]=targeting   

print(len(df))
df=df.loc[(df['Impressions'] != 0) & (df['Spend'] != 0)  &  (df['Clicks'] != 0)]
print(len(df))
```

```python
Y=pd.pivot_table(df.iloc[:, 1:],index=["Tipo"],values=["Clicks","Impressions","Spend","7 Day Total Sales ",
                                                                                                          "7 Day Total Orders (#)"],
                 aggfunc={"Clicks":np.sum,"Impressions":np.sum,"Spend":np.sum,
                          "7 Day Total Sales ":np.sum,"7 Day Total Orders (#)":np.sum,}) 

Acos=[]
CTR=Y["Clicks"]/Y["Impressions"]
CPC=(Y["Spend"]/Y["Clicks"])
ROAS=Y["7 Day Total Sales "]/Y["Spend"]
CR=Y["7 Day Total Orders (#)"]/Y["Clicks"]
Orders= Y["7 Day Total Orders (#)"]

for i in range(len(Y)):
    x=Y.iloc[i][1]
    y=Y.iloc[i][4]
    if x==0:
        u=0
        Acos.append(u)
    else:
        u=1/(x/y)
        Acos.append(u)



yf = pd.DataFrame()



yf['CTR'] = CTR
yf['CPC'] = CPC
yf['ROAS'] = ROAS
yf['CR'] = CR
yf['Orders'] = Orders
yf["Acos"]= Acos
yf.reset_index(inplace=True)
Y.reset_index(inplace=True)
yf["Impressions"]=Y["Impressions"]
yf["Sales"]= Y["7 Day Total Sales "]

yf["Clicks"]=Y["Clicks"]
yf["Spend"]=Y["Spend"]


yf=yf[["Tipo","Spend","Sales","Orders","CPC","CTR","CR","ROAS","Clicks","Impressions",]]

print(len(yf))
yf = yf[yf['CPC'].notna()]
print(len(yf))
```
